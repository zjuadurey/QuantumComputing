```python
import qiskit
import qiskit_qasm3_import

program = """
OPENQASM 3.0;
include "stdgates.inc";
qubit q;
h q;
"""

qc = qiskit_qasm3_import.parse(program)
print(qc)
```

```python
program2 = """
OPENQASM 3;
include "stdgates.inc";

qubit q;
bit c;

h q;
rz(pi / 4) q;
c = measure q;
if (c) {
  x q;
}
"""

qc2 = qiskit_qasm3_import.parse(program2)
print(qc2)
```

```python
program3 = """
OPENQASM 3;
defcalgrammar "openpulse";

cal {
  extern port d0;
  frame f = newframe(d0, 5.0e9, 0.0);
  play(f, gaussian(0.2 + 0.0im, 32ns, 8ns));
  shift_phase(f, pi / 2);
  delay[16ns] f;
}
"""
qc3 = qiskit_qasm3_import.parse(program3)
print(qc3)
```

```python
import openqasm3

program3 = """
OPENQASM 3;
defcalgrammar "openpulse";

qubit[1] q;
bit[1] c;

// 这里先放你的 pulse 代码
"""

tree = openqasm3.parse(program3)

for stmt in tree.statements:
    print(type(stmt).__name__)
```
